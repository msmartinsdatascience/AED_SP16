### Problema SR16 - Numerical 3-Dimensional Matching<br> 
$Instance$: Disjoint sets $W$, $X$ and $Y$, each cotaining $m$ elements, a size $s(a) \in  Z^+$ for each element $ a \in W \cup X \cup Y$, and a bound $ B \in Z^+$.<br>
$Question$: Can $W \cup X \cup Y$  be partitioned into $m$ disjoint sets $A_1, A_2, ..., A_m$ such that each $A_i$, contains exactly one element from each of $W$, $X$ and $Y$ and such that, for $1 \le i \le m$, $\sum_{a \in A_i}s(a)=B$?<br>
$Reference:$ [Garey and Jonhson, ----]. Transformation from 3DM <br><br>
Fonte: Computers and Intractability: A Guide to the Theory of NP-Completeness, página 224.


```python
import pandas as pd
import numpy as np
import time
```

```python
# Define a função do size
def peso(x):
    return x
    
def DefineConjunto():
    # Especifica conjunto de entrada
    x=int(input("Informe gerador de S: 1 -Sequência, \
    2 - Aleatório, 3 - Entre com seu conjunto: "))
    if x==3:
        listainput=input("Informe um conjunto de números separado por espaço: ").split()
        listainput=[int(string) for string in listainput]
        #  listainput=np.unique([int(string) for string in listainput])
        m=len(listainput) // 3
        S=listainput[:m*3]
        lenA=len(S)

    else:
        # número de elementos de cada um dos conjuntos W, X e Y.
        m=int(input("Informe o valor m, número de elementos de cada um dos conjuntos W, X e Y: "))
        lenA=3*m # define comprimento do conjunto A
        #  rng = np.random.RandomState(549351)
        if x==1:
            S=list(range(1,lenA+1))
        else: # aleatório
            aleatorio=np.unique(np.random.randint(1, lenA*200, lenA*100))
            S= aleatorio.tolist()[:lenA]

      #  S=rng.randint(1, 20, lenA).tolist()
    W, X, Y = S[:m], S[m:2*m], S[2*m:3*m] # avalia comprimento de W, X e Y:
    print("Conjunto A: ",S[:20],"...")
    print("Subconjuntos: \n W=",W[:10],"... \n X=", X[:10],"... \n Y=", Y[:10],"... ")

    return W, X, Y, m

def gerador_SP16(setA, setB, setC):
    # gera coleção de elementos X, Y, Z
    eventos=[]
    eventos_s=[]
    for a in setA:
      for b in setB:
        for c in setC:
          elemento=[peso(a),peso(b),peso(c)]
          eventos.append(elemento)
          eventos_s.append(sum(elemento))
    print(len(eventos))
    #print(eventos)

    #print(eventos_s)

    # Obtém valores de B que determinam m partições de 3 elementos
    df = pd.DataFrame(eventos_s, columns= ['total'])
    dicionario=df['total'].value_counts().to_dict()
    SolucaoB=[]
    for chave, valor in dicionario.items():
      if valor==m:
        SolucaoB.append(chave)
    if SolucaoB != []:
        print("Valore(s) de B para possível solução:",SolucaoB)
    else:
        print("O conjunto de entrada não pode ser particionado em",m, "subconjuntos de mesma soma.")
    return eventos, SolucaoB

def Solucao_SP16(SolucaoB=[]):

    if SolucaoB==[]:
        return print("")
    else:
        # Gera coleções de soma constante.
        contadorS=0
        for B in SolucaoB:
#            print("")
            A2=set()
            A3=[]
            for evento in eventos:
                if sum(evento)==B:
                    A2.update(set(evento))
                    A3.append(set(evento))
            #        print("Colecao soma",B, ":", evento)
            if A2==set(W)|set(X)|set(Y):
                contadorS=contadorS + 1
                print("Conjunto é uma partição de soma B=", B,".")
                print(A3)
        if contadorS==0:
            print("O conjunto não pode ser particionado em", m, "subconjuntos de soma", B, ".")

def DefineConjuntoAleatorio(m=3):
    lenA=3*m
    aleatorio=np.unique(np.random.randint(1, lenA*200, lenA*100))
    S= aleatorio.tolist()[:lenA]

      #  S=rng.randint(1, 20, lenA).tolist()
    W, X, Y = S[:m], S[m:2*m], S[2*m:3*m] # avalia comprimento de W, X e Y:
    print("Conjunto A: ",S[:20])
    #print("Subconjuntos: \n W=",W[:10],"... \n X=", X[:10],"... \n Y=", Y[:10],"... ")

    return W, X, Y, m
```

### Executar caso a caso

```python
# Chama funções
hst_tempo=[]
for i in range(1):
    print("="*80)
    W, X, Y, m = DefineConjunto()
    t0 = time.time()
    eventos, Solucao = gerador_SP16(W, X, Y)
    Solucao_SP16(Solucao)
    tempo=round(time.time() - t0,4)
    hst_tempo.append(tempo)
    print("Tempo estimado de execução:", round(time.time() - t0,6), "segundos")
print(hst_tempo)
```

```python
for item in hst_tempo:
  print(item)
```
